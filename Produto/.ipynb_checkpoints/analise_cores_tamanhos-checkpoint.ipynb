# Análise de Cores e Tamanhos de Roupa entre Países

Neste notebook, extraímos as cores e os tamanhos dos produtos a partir do seu nome (SKU) para identificar preferências regionais e explorar como o perfil de consumo (cor favorita, tamanho médio) varia de país para país.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', font_scale=1.0)

# 1. Carregar os dados
df = pd.read_csv('../data/sales_data_clean.csv')

# 2. Extrair Cor e Tamanho do Nome do Produto (Product_Name)
# Cores comuns nos nomes dos produtos
colors_pattern = r'\b(Black|White|Red|Blue|Green|Yellow|Orange|Gray|Pink|Purple|Brown)\b'
df['Color'] = df['Product_Name'].str.extract(colors_pattern, expand=False)

# Tamanhos de roupa comuns
sizes_pattern = r'\b(XS|S|M|L|XL|XXL)\b'
df['Size'] = df['Product_Name'].str.extract(sizes_pattern, expand=False)

# Ordenar tamanhos para os gráficos
size_order = ['XS', 'S', 'M', 'L', 'XL', 'XXL']
df['Size'] = pd.Categorical(df['Size'], categories=size_order, ordered=True)

print(f"Total de vendas: {len(df):,}")
print(f"Vendas com Cor identificada: {df['Color'].notna().sum():,}")
print(f"Vendas com Tamanho identificado: {df['Size'].notna().sum():,}")
```

## 1. Preferência de Cores por País

Vamos analisar qual a cor de produto mais vendida por país em volume de unidades (Orders).

```python
# Filtrar apenas produtos com cor
df_color = df[df['Color'].notna()]

# Contar vendas por País e Cor
color_country = df_color.groupby(['Country', 'Color'])['Sales_Order_ID'].count().reset_index(name='Orders')

# Calcular a percentagem que cada cor representa no total de vendas do país
country_totals = color_country.groupby('Country')['Orders'].transform('sum')
color_country['Share_pct'] = (color_country['Orders'] / country_totals) * 100

# Pivot table para o Heatmap
pivot_color = color_country.pivot(index='Country', columns='Color', values='Share_pct').fillna(0)

# Mapeamento das cores reais para as colunas do heatmap (se possível usar cmap neutro)
plt.figure(figsize=(12, 7))
sns.heatmap(pivot_color, annot=True, fmt='.1f', cmap='Blues', cbar_kws={'label': '% do total de vendas de cores no país'})
plt.title('Distribuição de Cores Vendidas por País (%)', fontweight='bold', fontsize=14)
plt.ylabel('País')
plt.xlabel('Cor')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top Cor por País
top_color_per_country = color_country.loc[color_country.groupby('Country')['Orders'].idxmax()]
print("Cor mais vendida por País:")
display(top_color_per_country[['Country', 'Color', 'Share_pct']].set_index('Country').sort_values('Share_pct', ascending=False).round(1))
```

## 2. Distribuição de Tamanhos de Roupa por País

Aqui consideramos apenas a Categoria Principal `Clothing` (Roupa) e os tamanhos identificados (XS, S, M, L, XL, XXL).

```python
# Filtrar apenas produtos de Roupa e com Tamanho
df_clothing = df[(df['Main_Category'] == 'Clothing') & (df['Size'].notna())]

# Contar vendas por País e Tamanho
size_country = df_clothing.groupby(['Country', 'Size'])['Sales_Order_ID'].count().reset_index(name='Orders')

# Calcular a percentagem de cada tamanho dentro do país
country_size_totals = size_country.groupby('Country')['Orders'].transform('sum')
size_country['Share_pct'] = (size_country['Orders'] / country_size_totals) * 100

# Pivot table para visualização empilhada (Stacked Bar Chart)
pivot_size = size_country.pivot(index='Country', columns='Size', values='Share_pct').fillna(0)
pivot_size = pivot_size[size_order] # Garantir a ordem XS -> XXL

# Heatmap para Tamanhos
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_size, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': '% de vendas de roupa no país'})
plt.title('Distribuição de Tamanhos de Roupa por País (%)', fontweight='bold', fontsize=14)
plt.ylabel('País')
plt.xlabel('Tamanho')
plt.tight_layout()
plt.show()

# Gráfico de Barras Empilhadas
ax = pivot_size.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='viridis', edgecolor='white')
plt.title('Share de Tamanhos de Roupa Vendidos por País (Stacked)', fontweight='bold', fontsize=14)
plt.ylabel('Percentagem (%)')
plt.xlabel('País')
plt.legend(title='Tamanho', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')

# Adicionar os valores percentuais dentro das barras
for c in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 3 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center', fontsize=8, color='white', fontweight='bold')

plt.tight_layout()
plt.show()
```
