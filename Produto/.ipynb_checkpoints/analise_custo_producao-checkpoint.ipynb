# Análise de Custos de Produção Médios

Neste notebook, exploramos o custo médio de produção (Production Cost) ao longo do tempo (Ano) e por Região. A análise divide-se em três níveis de detalhe: Categoria Principal, Subcategoria e Produto (SKU).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib.ticker as mticker

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', font_scale=1.0)

# 1. Carregar os dados
df = pd.read_csv('../data/sales_data_clean.csv')

print(f"Total de vendas: {len(df):,}")
print(f"Média global de Custo de Produção: €{df['Production_Cost'].mean():.2f}")
```

## 1. Custo Médio por Categoria Principal, Região e Ano

```python
# Agrupar por Categoria Principal, Região e Ano
cost_cat = df.groupby(['Main_Category', 'Region', 'Year'])['Production_Cost'].mean().reset_index(name='Avg_Cost')

g = sns.FacetGrid(cost_cat, col='Region', hue='Main_Category', height=5, aspect=1.2)
g.map(sns.lineplot, 'Year', 'Avg_Cost', marker='o', linewidth=2)
g.add_legend(title='Categoria Principal')
g.set_axis_labels("Ano", "Custo Médio (€)")
g.set_titles(col_template="{col_name}", fontweight='bold')
for ax in g.axes.flatten():
    ax.xaxis.set_major_locator(mticker.MaxNLocator(integer=True))

plt.subplots_adjust(top=0.85)
g.fig.suptitle('Evolução do Custo Médio por Categoria Principal e Região', fontweight='bold', fontsize=14)
plt.show()

# Heatmap para tabela de dados
pivot_cat = cost_cat.pivot_table(index=['Main_Category', 'Region'], columns='Year', values='Avg_Cost')
print("Tabela: Custo Médio por Categoria e Região ao longo dos anos (€)")
display(pivot_cat.round(2))
```

## 2. Custo Médio por Subcategoria, Região e Ano

Como existem muitas subcategorias, mostramos os gráficos divididos por Região.

```python
# Agrupar por Subcategoria, Região e Ano
cost_subcat = df.groupby(['Subcategory', 'Region', 'Year'])['Production_Cost'].mean().reset_index(name='Avg_Cost')

regions = cost_subcat['Region'].dropna().unique()

for region in regions:
    d = cost_subcat[cost_subcat['Region'] == region]
    
    plt.figure(figsize=(14, 8))
    pivot = d.pivot(index='Subcategory', columns='Year', values='Avg_Cost').fillna(0)
    sns.heatmap(pivot, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Custo Médio (€)'})
    plt.title(f'Evolução do Custo Médio por Subcategoria - {region}', fontweight='bold')
    plt.ylabel('Subcategoria')
    plt.xlabel('Ano')
    plt.tight_layout()
    plt.show()
```

## 3. Custo Médio por Produto, Região e Ano

Sendo que há milhares de produtos, vamos identificar os **Top 15 Produtos com o Custo de Produção Médio mais elevado** em cada região e observar a sua variação temporal.

```python
# Agrupar por Produto, Região e Ano
cost_prod = df.groupby(['Product_Name', 'Region', 'Year'])['Production_Cost'].mean().reset_index(name='Avg_Cost')

for region in regions:
    d = cost_prod[cost_prod['Region'] == region]
    
    # Descobrir os Top 15 produtos mais caros (em média, agregando todos os anos)
    top_expensive = d.groupby('Product_Name')['Avg_Cost'].mean().nlargest(15).index
    d_top = d[d['Product_Name'].isin(top_expensive)]
    
    plt.figure(figsize=(14, 8))
    pivot = d_top.pivot(index='Product_Name', columns='Year', values='Avg_Cost').fillna(0)
    sns.heatmap(pivot, annot=True, fmt='.0f', cmap='Reds', cbar_kws={'label': 'Custo Médio (€)'})
    plt.title(f'Custo Médio dos 15 Produtos Mais Caros - {region}', fontweight='bold')
    plt.ylabel('Produto (SKU)')
    plt.xlabel('Ano')
    plt.tight_layout()
    plt.show()
```
